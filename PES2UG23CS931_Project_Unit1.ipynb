```python
from transformers import pipeline
```

```python
qa_bot = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

print("✅ QA model loaded successfully")
```

```python
policy_text = """
Our company offers a 30-day return policy for all products.
Customers must return items in their original packaging.
Refunds are processed within 7 business days after receiving the returned item.
Shipping charges are non-refundable.
For technical support, customers can contact us via email or phone.
Customer support is available from Monday to Friday, 9 AM to 6 PM.
"""
```

```python
def ask_faq(question):
    result = qa_bot({
        "question": question,
        "context": policy_text
    })
    return result["answer"]

```

```python
questions = [
    "What is the return policy?",
    "How long does it take to get a refund?",
    "Are shipping charges refundable?",
    "When is customer support available?"
]

for q in questions:
    print(f"Q: {q}")
    print(f"A: {ask_faq(q)}\n")
```
